## All detectors rpeak databases feature extraction

```python
import pandas as pd
import os
import numpy as np
from ecgdetectors import Detectors
from time import time as time
import seaborn as sns
from hrvanalysis import remove_outliers, remove_ectopic_beats, interpolate_nan_values, get_time_domain_features, get_geometrical_features, get_frequency_domain_features, get_csi_cvi_features, get_poincare_plot_features, get_sampen
from time import time as time
import warnings
warnings.filterwarnings('ignore')
```

```python
path = '/hdd/physio/vf/finaldfs/'
fs = 250
```

```python
df = pd.read_csv(path + 'finalrpeak5secpt.csv', index_col = 0)
df
```

```python
%%time
def featureExtraction(df, fs):
    df1 = df.vf.copy()
    df.drop(columns = ['vf'], inplace = True)
    final = list()
    for i in range(len(df)):
        peaks = []
        peaks = df.loc[i][df.loc[i] == 1].index.values
        peaks = [int(x) for x in peaks]
        if (len(peaks) == 0 or len(peaks) == 1):
            continue
        vf = df1.loc[i]
        rr = np.diff(peaks) / fs 
        hr = np.diff(peaks) / fs 
        hr = 60 / rr
        rmssd = np.sqrt(np.mean(np.square(rr)))
        sdnn = np.std(rr)
        mean_rr = np.mean(rr)
        mean_hr = np.mean(hr)
        std_hr = np.std(hr)
        min_hr = np.min(hr)
        max_hr = np.max(hr)
        time_domain_features = {'RMSSD': rmssd, 'SDNN': sdnn, 'MEAN_RR': mean_rr, 'MEAN_HR': mean_hr, 'STD_HR': std_hr, 'MIN_HR': min_hr, 'MAX_HR': max_hr}
        frequency_domain_features = get_frequency_domain_features(peaks)
        features = [time_domain_features, frequency_domain_features]
        del peaks
        new = {}
        for x in features:
            for y in x:
                new[y] = x[y]

        single_window_df = pd.DataFrame(list(new.items()))
        single_window_df = single_window_df.transpose()
        columns = single_window_df.iloc[0]
        single_window_df = single_window_df[1:]
        single_window_df.columns = columns
        single_window_df.drop(columns = ['lfnu', 'hfnu'], inplace = True)
        single_window_df.insert(0, 'vf', vf)
        final.append(single_window_df)
        del single_window_df
    return pd.concat(final)

finaldf = featureExtraction(df, fs)
finaldf.reset_index(drop = True, inplace = True)
```

```python
finaldf = finaldf.dropna(how='any')
finaldf
```
